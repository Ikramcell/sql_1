```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

print("TensorFlow version:", tf.__version__)  # Should print something like 2.17.0 or higher
```

```python
# Load the MNIST dataset
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Check shapes
print("Training images shape:", train_images.shape)  # (60000, 28, 28)
print("Training labels shape:", train_labels.shape)  # (60000,)
print("Test images shape:", test_images.shape)      # (10000, 28, 28)
```

```python
# Plot the first 25 images
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])  # No ticks
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)  # Show image in black/white
    plt.xlabel(train_labels[i])  # Label below
plt.show()
```

```python
# Normalize pixel values to 0-1 range
train_images = train_images / 255.0
test_images = test_images / 255.0

# Check a sample pixel value
print("Sample pixel after normalization:", train_images[0][14][14])  # Should be between 0 and 1
```

```python
# Build the model
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),  # Flatten 28x28 image to 784 pixels
    tf.keras.layers.Dense(128, activation='relu'),  # Hidden layer with 128 neurons, ReLU activation
    tf.keras.layers.Dense(10)                       # Output layer with 10 neurons (one per digit 0-9)
])

# Summary of the model
model.summary()
```

```python
model.compile(
    optimizer='adam',  # Good default optimizer
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),  # For multi-class classification
    metrics=['accuracy']  # Track % correct
)
```

```python
# Train for 10 epochs
history = model.fit(train_images, train_labels, epochs=10)
```

```python
# Evaluate on test data
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print("\nTest accuracy:", test_acc)  # Should be around 0.88 or higher
```

```python
# Predict on first test image
probability_model = tf.keras.Sequential([model, tf.keras.layers.Softmax()])  # Convert logits to probabilities
predictions = probability_model.predict(test_images[:5])

# Show results
for i in range(5):
    plt.figure(figsize=(2,2))
    plt.imshow(test_images[i], cmap=plt.cm.binary)
    plt.title(f"Predicted: {np.argmax(predictions[i])}, Actual: {test_labels[i]}")
    plt.show()
```
